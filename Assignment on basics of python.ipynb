### Exercise 1: Prime Numbers

```python
# Write a Python program that checks whether a given number is prime or not.
# A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
```

```python
number = int(input("Enter the number to check for prime number"))
if number > 1:
   for i in range(2, number):
       print(i)
       if number % i == 0:
           print("Number is not prime")
           break
       else:
            print("Number is prime")
            break
else:
    print("Number is not prime")
```

```python
# Exercise 2: Product of Random Numbers
# Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. 
# The program should then check if the user's answer is correct and display an appropriate message.
```

## Exercise 3: Squares of Even/Odd Numbers

```python
# Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200.
# Choose either even or odd numbers and document your choice in the code.
```

```python
choice = input("you want know the squares of even or odd")
if choice == 'even':
    for i in range(100, 200):
        if i % 2 == 0:
            print("Even number is", i)
            print("Squares of ", i, 'is', i**2)
elif choice == 'odd':
    for i in range(100, 200):
        if i % 2 == 1:
            print("Odd number is", i)
            print("Squares of", i, 'is', i**2)
```

## Excercise 4  Hospital billing system

```python
'''Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400'''
```

```python
print("------- Billing Criteria-------")
print("Room charge per day is 2,000")
print("Doctor consultation fees are 1,500")
print("Lab test charges: Based on the number of tests taken. Each test costs ₹300")
print("Medicine charges: Total cost of medicines provided")
print("Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied")
Stay = 4
lab_test = 3
medicine_charges = 2400
total_roomcharges = 2000*Stay
lab_testcharge = lab_test*300
total_bill = total_roomcharges  + lab_testcharge  + medicine_charges
print("Total bill is", total_bill)
if total_bill>10000:
    final_bill = total_bill - total_bill*(10/100)
    print("With 10% discount your final bill is", final_bill)
```
